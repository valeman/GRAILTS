We define big datasets as datasets having more than 1000 test queries.

```python
big_datasets = df[df.TEST_size >= 1000]

big_datasets_1NN = big_datasets[big_datasets.NN == 1]

big_datasets_1NN
```

Speedup is GRAILtime/optimizedtime. 

```python
speedup_opq64 = big_datasets_1NN['GRAIL exact time']/big_datasets_1NN['OPQ (64) time']
speedup_opq128 = big_datasets_1NN['GRAIL exact time']/big_datasets_1NN['OPQ (128) time']
speedup_pq = big_datasets_1NN['GRAIL exact time']/big_datasets_1NN['PQ time']

speedup_opq64
```

```python
import matplotlib.pyplot as plt 
plt.scatter(big_datasets_1NN['TEST_size'], speedup_pq)
plt.xlabel("Number of Test queries")
plt.ylabel("Speedup of PQ")
```

```python
plt.scatter(big_datasets_1NN['TRAIN_size'], speedup_pq)
plt.xlabel("TRAIN size")
plt.ylabel("Speedup of PQ")
```

Training size seems to give a better estimation of the speedup. The speedup seems to be logarithmic with respect to the number of training points. 

```python
plt.scatter(big_datasets_1NN['TRAIN_size'], speedup_opq128)
plt.xlabel("TRAIN size")
plt.ylabel("Speedup of OPQ (128)")
```

```python
plt.scatter(big_datasets_1NN['TRAIN_size'], speedup_opq64)
plt.xlabel("TRAIN size")
plt.ylabel("Speedup of OPQ (128)")
```
