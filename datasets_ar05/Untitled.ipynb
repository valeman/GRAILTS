```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.stattools import adfuller

```

```python
ts = np.load('series2_200.npy')

x = np.arange(ts[0].shape[0])
plt.plot(x, ts[0], c='blue')
plt.show()
```

```python

def adf_test(x, pval = 0.05):
    """
    Test if the time series is stationary
    :param x: time series
    :return:
    """
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(x, autolag='AIC')
    if dftest[1] < pval:
        return True
    else:
        return False
    
def make_stationary(x):
    m = x.shape[0]
    for i in range(1,m):
        x[i] -= x[i-1]

def preprocess(TS):
    n,m = TS.shape
    for x in TS:
        if not adf_test(x):
            make_stationary(x)
```
